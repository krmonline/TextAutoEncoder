```python
import tensorflow as  tf
import numpy as np
import pythainlp
import deepcut
import pythainlp.word_vector
import pandas as pd
```

```python
model = pythainlp.word_vector.get_model()
```

```python
thai2dict = {}
thaiword2Index = {}
i = 0
thaiIndex2word  = []
embedding_matrix = np.zeros([len(model.index2word),300])
for word in model.index2word:
    thai2dict[word] = model[word]
    embedding_matrix[i] = model.get_vector(word)
    thaiIndex2word.append(word)
    thaiword2Index[word] = i
    i = i + 1
thai2vec = pd.DataFrame.from_dict(thai2dict,orient='index')
vocab_size = len(thai2vec)
thai2vec
```

```python
import glob
```

```python
pd.set_option('display.max_columns',None)
n = 0
arr_category = []
arr_details = []
data = pd.DataFrame(columns=['id','detail','category'])
#display(data)
for category in ['crime','politic','world']:
    for i in glob.glob("/Users/chakritphain/Downloads/sanook1/" + category +"/*/content.txt"):
        #print(i)
        news_id = i.split("/")[-2]
        dftmp = pd.read_csv(i,names='c')
        #display(dftmp)
        arr_tmp = dftmp['c'].values
        result = ' '.join(map(str,(arr_tmp)))
        data = data.append({"id":news_id,"detail":result,"category":category},ignore_index=True)
display(data)
```

```python
def CheckNull(x):
    if pd.isnull(x):
        x = "Unknow"
    return x
def tokenize(x):
    return ",".join(pythainlp.word_tokenize(x))
def string2vec(x):
    arr = pythainlp.word_tokenize(x)
    tmp = []
    for word in arr:
        if word in model.vocab:
            tmp.append(thaiword2Index[word])
        else:
            tmp.append(thaiword2Index[""])
    return tmp

data['detail'] = data['detail'].apply(CheckNull)
data['token'] = data['detail'].apply(tokenize)
```

```python
maxlen = 0
for i in data.token.values:
    tmp = i.split(",")
    l = len(tmp)
    if  maxlen < l:
        maxlen = l
print(maxlen)
```

```python
arrTokenW2V = []
for i in data.detail.values:
    arrTokenW2V.append(string2vec(i))
arrTokenW2V = np.array(arrTokenW2V)
arrTokenW2V.shape
```

```python
arrTokenW2V[0]
```

```python
X = tf.keras.preprocessing.sequence.pad_sequences(
    arrTokenW2V, maxlen=maxlen, dtype='object', padding='pre',
    truncating='pre', value=string2vec(" ")
)
X[0]
def padding(x):
    x = tf.keras.preprocessing.sequence.pad_sequences(
        np.array([x]), maxlen=10, dtype='object', padding='pre',
        truncating='pre', value=string2vec(" ")
    )
    return x[0]
```

```python
def split_sequence(sequence, n_steps):
    X = []
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        #print("frome",i,"to",str(end_ix))
        if end_ix > len(sequence)-1:
            break
        #print(sequence[i:end_ix])
        X.append(sequence[i:end_ix])
    return np.array(X)
```

```python
split_sequence(arrTokenW2V[0],10)
```

```python
seq_step  = 10
X = []
#print(X)
for i  in arrTokenW2V[:]:
    if len(i) < seq_step:
        continue
    if len(X):
        Xtmp = split_sequence(i,seq_step)
        X = np.append(X,Xtmp,axis=0)
    else:
        X = split_sequence(i,seq_step)
print(X.shape)
```

```python
#random
```

```python
history=[]
#del(model1)
model1 = tf.keras.Sequential()
model1.add(tf.keras.layers.Embedding(vocab_size,300,weights=[embedding_matrix], input_length=10)),
model1.add(tf.keras.layers.GRU(1, activation='relu', return_sequences=True))
model1.add(tf.keras.layers.GRU(5, activation='relu', return_sequences=False))
#model1.add(tf.keras.layers.Reshape((5,1)))
model1.add(tf.keras.layers.RepeatVector(1))
model1.add(tf.keras.layers.GRU(10, activation='relu', return_sequences=False))
#model1.add(tf.keras.layers.GRU(1, activation='tanh', return_sequences=True))
#model1.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(1)))
#
model1.add(tf.keras.layers.Reshape((10,1)))
#model1.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(1)))
#model1.add(tf.keras.layers.Flatten(100))

model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss='mse')
model1.summary()
model1.load_weights("textAE.f5")
```
